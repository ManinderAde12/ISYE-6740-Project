# PCA Data EDA

__Objective:__ Prepare some quick visual to help describe the baseline data.

## Import Packages And Data

```python
# packages
import pandas as pd
from plotnine import *
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data
base_data = pd.read_csv('Pnuemonia Images PCA Reduced.csv')\
    .drop('Unnamed: 0', axis = 1)
print('base_data shape:', base_data.shape)
base_data.head()
```

## Tidy And Transform

```python
df = base_data.melt(id_vars = ['y', 'set'])
# df = df[df['set'] == 'train']
df['y'] = df['y'].str.title()
df.head()
```

## Plot Density Ridges

```python
fig = (
    ggplot(df, aes("value")) + 
        geom_density(aes(fill = "y"), alpha = 0.5) + 
        facet_wrap("~ variable") + 
        labs(
            title = "Density Plots Of PCA Dimensions By Class",
            fill = "",
            x = "Value",
            y = "Density"
        ) + 
        scale_fill_manual(values = ["grey", "blue", "orange"]) + 
        theme_bw(base_size = 12) + 
        theme(
            figure_size = (8, 8),
            legend_position = "top",
            plot_title = element_text(hjust = 0.5, face = "bold", size = 14)
        )
)

fig.save(filename = 'images/PCA EDA Density Plots.png', dpi = 300)

fig
```

## Plot Two Dimensional Relationships

Reminder: Best components are [PCA1, PCA2, PCA3, PCA4, PCA9], so we want to show 10 pairings.

```python
import itertools
key_cols = ['pca1', 'pca2', 'pca3', 'pca4', 'pca9', 'y', 'set']
key_cols_combos = list(itertools.combinations(key_cols, 2))
df2 = base_data[key_cols]
```

```python
# create pair grid
sns.set(font_scale = 1.5)
sns.set_theme(style = 'whitegrid', palette = ['gray', 'blue', 'orange'])
g = sns.PairGrid(
    data = df2,
    hue = 'y'
)
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.add_legend()

g.fig.subplots_adjust(top = 0.9)
g.fig.suptitle('Pairwise Comparison Of Best Principal Components')
g.savefig('images/Pairwise Grid Of Best PCs.png', bbox_inches = 'tight', dpi = 300)

g
```
